3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT

```python
def Q1(g):
    return np.percentile(g,25)
```

```python
def Q2(g):
    return np.percentile(g,50)
```

```python
def Q3(g):
    return np.percentile(g,75)
```

```python
def std(g):
    return np.std(g)
```

```python
pd.pivot_table(data=df_tuyensinh,index="KT",values="DH1",aggfunc=["sum","mean","median","min","max",std,Q1,Q2,Q3])
```

2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
Gợi ý
- Trực quan hóa biểu đồ Box-plot, histogram và QQ-plot theo phân nhóm là giá trị của 'phanlopT1'

```python
# Box-plot theo nhóm
df_tuyensinh.boxplot(column='T1', by='phanlopt1')
```

```python
# Histogram theo nhóm
df_tuyensinh.groupby('phanlopt1')['T1'].plot(kind='hist', bins=15, alpha=0.7, legend=True)
```

```python
# QQ-Plot theo nhóm
import matplotlib.pyplot as plt
from scipy import stats

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, category in enumerate(df_tuyensinh['phanlopt1'].cat.categories):
    data = df_tuyensinh[df_tuyensinh['phanlopt1'] == category]['T1']
    stats.probplot(data, dist="norm", plot=axes[i])
    axes[i].set_title(f'Q-Q Plot - {category}')

plt.tight_layout()
plt.show()
```
