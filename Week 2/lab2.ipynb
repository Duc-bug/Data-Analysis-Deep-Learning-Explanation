1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần

```python
import pandas as pd
import numpy as np
```

```python
df_tuyensinh = pd.read_csv("processed_dulieuxettuyendaihoc.csv")
df_tuyensinh.head()
```

2. Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính

```python
df_tuyensinh.sort_values(by=['GT', 'DH2'], ascending=[True, True])
```

3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT

```python
def Q1(g):
    return np.percentile(g,25)
```

```python
def Q2(g):
    return np.percentile(g,50)
```

```python
def Q3(g):
    return np.percentile(g,75)
```

```python
def std(g):
    return np.std(g)
```

```python
pd.pivot_table(data=df_tuyensinh,index="KT",values="DH1",aggfunc=["sum","mean","median","min","max",std,Q1,Q2,Q3])
```
